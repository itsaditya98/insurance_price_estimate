```python
import opendatasets as od
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

imported libraries for data analysis

```python
medical_df = pd.read_csv('medical.csv')
```

```python
medical_df
```

explored the data

```python
input_cols=['age', 'sex', 'bmi', 'children', 'smoker', 'region',]
input_df = medical_df[input_cols].copy()
```

```python
input_df
```

Data Preprocessing and feature engineering before training model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
numeric_cols = ['age', 'bmi', 'children']
scaler = StandardScaler()
scaler.fit(input_df[numeric_cols])
```

```python
numeric_cols
```

```python
input_df[numeric_cols]= scaler.transform(input_df[numeric_cols])
```

```python
input_df
```

```python
cat_cols = ['sex', 'smoker']

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoder.fit(input_df[cat_cols])
```

```python
encoder.categories_
```

```python
encoded_cols = list(encoder.get_feature_names_out(cat_cols))
```

```python
encoded_cols
```

```python
input_df[encoded_cols] = encoder.transform(input_df[cat_cols])
```

```python
input_df
```

```python
target = medical_df.charges
target
```

```python
input = input_df[encoded_cols+numeric_cols]
```

```python
input
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs_train, inputs_test, targets_train, targets_test = train_test_split(input, target, test_size=0.1)
```

```python
inputs_train
```

```python
def rmse(targets, predictions):
    return np.sqrt(np.mean(np.square(targets - predictions)))
```

training model using Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create and train the model
model = LinearRegression().fit(inputs_train, targets_train)

# Generate predictions
predictions_train = model.predict(inputs_train)

# Compute loss to evalute the model
loss = rmse(targets_train, predictions_train)
print('Training Loss:', loss)

# Generate predictions
predictions_test = model.predict(inputs_test)

# Compute loss to evalute the model
loss = rmse(targets_test, predictions_test)
print('Test Loss:', loss)
```

running on single input

```python
new_input={'age' : 27,
           'sex': 'male', 'bmi' : 32, 'children' : 0, 'smoker' : 'yes', 'region' : 'northwest'}
```

```python
def predict_input(single_input):
    input_df = pd.DataFrame([single_input])
    input_df[numeric_cols] = scaler.transform(input_df[numeric_cols])
    input_df[encoded_cols] = encoder.transform(input_df[cat_cols])
    X_input = input_df[encoded_cols + numeric_cols]
    pred = model.predict(X_input)[0]
    return pred
```

```python
predict_input(new_input)
```

now saving the model

```python
insurance={'model': model,
    'scaler': scaler,
    'encoder' : encoder,
    'input_cols': input_cols,
    'target': target,
    'numeric_cols': numeric_cols,
    'cat_cols': cat_cols,
    'encoded_cols': encoded_cols
}
```
